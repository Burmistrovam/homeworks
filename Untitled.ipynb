```python
from pymystem3 import Mystem

text = """На бледно-голубой эмали,
Какая мыслима в апреле,
Березы ветви поднимали
И незаметно вечерели"""

m = Mystem()
lemmas = m.lemmatize(text)

print(type(lemmas)) # функция возвращает список
print(''.join(lemmas))
```
